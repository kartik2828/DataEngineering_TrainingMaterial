1. Reverse a string

Write a function that takes a string and returns the string reversed.

```python
Str = 'Apple'
def rev_string(s):
    return s[::-1]
obj = rev_string(Str)
print(obj)
```

2. Check if a string is a palindrome

Determine whether a given string reads the same forwards and backwards.

```python
str = 'NitiN'
def pallindrome(s):
    return s  == s[::-1]
obj1  = pallindrome(str)
print(obj1)
```

3. First non-repeating character

Find the first character in a string that does not repeat. Return its index or -1 if none exists.

```python
str = 'HHeeloo'
def non_repeating(s):
    alpha = {}
    for i in s:
        if i in alpha:
            alpha[i] += 1
        else:
            alpha[i] = 1
    for idx, ch in enumerate(s):
        if alpha[ch] == 1:
            return idx
    return -1
obj2 = non_repeating(str)
print(obj2)
```
