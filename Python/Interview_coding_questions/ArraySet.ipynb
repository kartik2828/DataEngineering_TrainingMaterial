# Python Array Interview Questions

10 prioritized array problems for a Data Engineer (3 years experience). Each question is followed by an empty code cell for your solution. Questions are ordered by likelihood of appearing in interviews (highest first).

1. Two Sum

Given an array of integers and a target value, return indices of the two numbers such that they add up to the target.

2. Product of Array Except Self

Given an array nums, return an array answer such that answer[i] is the product of all the elements of nums except nums[i]. Solve without using division and in O(n) time.

3. Three Sum

Given an array nums, find all unique triplets in the array which gives the sum of zero.

4. Maximum Subarray (Kadane's Algorithm)

Find the contiguous subarray (containing at least one number) which has the largest sum and return its sum.

5. Move Zeroes

Given an array nums, write a function to move all 0's to the end while maintaining the relative order of the non-zero elements (in-place).

6. Remove Duplicates from Sorted Array

Given a sorted array nums, remove the duplicates in-place such that each element appears only once and return the new length.

7. Contains Duplicate / Find Duplicate

Given an array of integers, determine if any value appears at least twice in the array. Optionally, find the duplicate number when constraints apply.

8. Merge Sorted Arrays (in-place)

Given two sorted integer arrays nums1 and nums2, merge nums2 into nums1 as one sorted array. Assume nums1 has enough space at the end to hold additional elements.

9. Search in Rotated Sorted Array

Suppose an array sorted in ascending order is rotated at some pivot. Given target, search in O(log n) time and return its index or -1 if not found.

10. Find Missing Number

Given an array containing n distinct numbers taken from 0, 1, 2, ..., n, find the one that is missing. Aim for O(n) time and O(1) extra space.
