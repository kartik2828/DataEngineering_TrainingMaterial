- Find the second largest element in an array

```python
arr = [1,2,4,34,65,21,34,56,78,90,12,34,56,78]
def largest_element(arr):
    max_element = arr[0]
    secondMax = arr[0]-1
    for i in arr:
        if i>max_element:
            max_element = i
        for j in arr:
            if j>secondMax and j<max_element:
                secondMax = j
    return secondMax
obj = largest_element(arr)
print(obj)            

     
```

- Rotate an array by k positions to the  <br>

arr = [1, 2, 3, 4, 5],  k = 2 <br>

Output → [4, 5, 1, 2, 3]

```python
def rotate_array(arr, k):
    part1 = arr[-k:]
    part2 = arr[:-k]
    arr = part1 + part2
    return arr
obj = rotate_array([1,2,3,4,5], 2)
print(obj)    
```

- Find all pairs in an array that sum to a target <br>

arr = [1, 2, 3, 4, 5],  target = 5 <br>

Output → (1, 4), (2, 3)

```python
def array_pairs(arr, target):
    pairs = []
    for i in range(len(arr)):
        for j in range(i+1, len(arr)):
            if arr[i] + arr[j] == target:
                pairs.append((arr[i], arr[j]))
    return pairs
obj  = array_pairs([1,2,3,4,5],5)
print(obj)            
```

- Move all zeroes to the end without changing order <br>

arr = [0, 1, 0, 3, 12, 0, 5]<br>
Output → [1, 3, 12, 5, 0, 0, 0]

```python
def move_zeroes(arr):
    non_zeroes = []
    zeroes = []
    for i in arr:
        if i==0:
            zeroes.append(i)
        else:
            non_zeroes.append(i)    
    return non_zeroes + zeroes
obj = move_zeroes([0,1,0,3,12,0,5])   
print(obj)     
```

- Find the missing number in array 1 to N <br>

arr = [1, 2, 4, 5, 6],  N = 6<br>

Output → 3  (missing number)

```python
def missing_number(arr):
    n = 6
    expected_sum = n*(n+1)//2 
    actual_sum = 0
    for i in arr:
        actual_sum += i
    return expected_sum - actual_sum
obj = missing_number([1,2,4,5,6])
print(obj)
```

- Check if array contains duplicate within k distance <br>

arr = [1, 2, 3, 1, 4, 5],  k = 3 <br>

Output → True  (1 appears at index 0 and index 3, distance = 3)

```python
def duplicate_number(arr, k):
    for i in range(len(arr)):
        for j in range(i+1, len(arr)):
            if arr[j] == arr[i]:
                if j - i <= k:
                    return True
    return False
obj = duplicate_number([1,2,3,1,4,5], 3)  
print(obj)            
            
```

- Find maximum product of two integers in array <br>

arr = [1, 2, 3, 4, 5] <br>
Output → 20  (4 × 5 = 20)

```python
def two_int_product(arr):
    largest = 0
    second_largest = 0
    for i in arr:
        if i>largest:
            second_largest = largest
            largest= i
    return second_largest * largest
obj = two_int_product([1,2,3,4,5])
print(obj)        

```
