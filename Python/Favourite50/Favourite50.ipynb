```python
"""
Given a string s, find the first non-repeating character and return its index.
If it doesn’t exist, return -1.

Input: "leetcode"
Output: 0

Input: "loveleetcode"
Output: 2

Input: "aabb"
Output: -1

"""

# Brutefoce
def my_func(s: str) -> int:
    count = {}
    for char in s:
        count[char] = count.get(char,0) + 1 
    
    for index, char in enumerate(s):
        if count[char] == 1:
            return index
    return -1

obj = my_func("leetcode")    
print(obj)


#optimize 
from collections import Counter

def firstUniqChar(s: str) -> int:
    freq = Counter(s)
    
    for i, ch in enumerate(s):
        if freq[ch] == 1:
            return i
    return -1
obj = firstUniqChar("loveleetcode")
print(obj)
```

```python
"""
Example:

Input: nums = [1, 1, 2]

Output: 2, nums = [1, 2, _]
"""

def func(nums: list):

    sorted = []

    for i in nums:
        if i not in sorted:
            sorted.append(i)

    return sorted

obj = func([1,1,2,3,3,4])   
print(obj)
```

```python
# Given a string s, find the length of the longest substring without repeating characters.
# Sliding WINDOW
# Input: "abcabcbb"
# Output: 3  # "abc"

def func(s: str)-> int:
    char_set = set()
    left = 0
    max_length = 0

    for right in range(len(s)):
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1
        char_set.add(s[right])
        max_length = max(max_length, right - left + 1)

    return max_length
obj = func("abcabcbb")
print(obj)
```

```python
# Sliding WINDOW
#Given a string s, find the length of the longest substring with at most 2 distinct characters.
# Input: "eceba"
# Output: 3   # "ece"

def longest_substring_k_distinct(s: str) -> int:
    freq = {}
    left = 0
    max_length = 0

    for right in range(len(s)):
        # expand window
        freq[s[right]] = freq.get(s[right], 0) + 1

        # shrink window if more than 2 distinct chars
        while len(freq) > 2:
            freq[s[left]] -= 1
            if freq[s[left]] == 0:
                del freq[s[left]]
            left += 1

        # update max length
        max_length = max(max_length, right - left + 1)

    return max_length
obj = longest_substring_k_distinct("eceba")
print(obj)
```

```python
# Two sum 

# Input: nums = [2,7,11,15], target = 9
# Output: [0,1]

def two_sum(nums:list, target: int) -> list:

    left = 0
    right = len(nums)-1

    while left<right:
        current_sum = nums[left] + nums[right]

        if current_sum == target:
            return [left, right]
        elif current_sum < target:
            left += 1
        else:
            right -= 1
    return []
obj = two_sum([2,7,11,15], 9)
print(obj)            
```
