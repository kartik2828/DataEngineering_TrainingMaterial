```python
"""
Given a string s, find the first non-repeating character and return its index.
If it doesn’t exist, return -1.

Input: "leetcode"
Output: 0

Input: "loveleetcode"
Output: 2

Input: "aabb"
Output: -1

"""

# Brutefoce
def my_func(s: str) -> int:
    count = {}
    for char in s:
        count[char] = count.get(char,0) + 1 
    
    for index, char in enumerate(s):
        if count[char] == 1:
            return index
    return -1

obj = my_func("leetcode")    
print(obj)


#optimize 
from collections import Counter

def firstUniqChar(s: str) -> int:
    freq = Counter(s)
    
    for i, ch in enumerate(s):
        if freq[ch] == 1:
            return i
    return -1
obj = firstUniqChar("loveleetcode")
print(obj)
```

```python
"""
Example:

Input: nums = [1, 1, 2]

Output: 2, nums = [1, 2, _]
"""

def func(nums: list):

    sorted = []

    for i in nums:
        if i not in sorted:
            sorted.append(i)

    return sorted

obj = func([1,1,2,3,3,4])   
print(obj)
```
